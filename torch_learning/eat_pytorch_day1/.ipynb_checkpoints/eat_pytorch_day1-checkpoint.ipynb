```python
def preprocessing(dfdata):
    
    dfresult = pd.DataFrame()

    # Pclass
    dfPclass = pd.get_dummies(dfdata['Pclass'])
    dfPclass.columns = ['Pclass_' + str(x) for x in dfPclass.columns]
    dfresult = pd.concat([dfresult, dfPclass], axis=1)
    '''
       Pclass_1  Pclass_2  Pclass_3
    0         0         0         1
    1         1         0         0
    2         0         0         1
    '''
    
    # Sex
    dfSex = pd.get_dummies(dfdata['Sex'])
    dfresult = pd.concat([dfresult, dfSex], axis=1)
    '''
	    Pclass_1	Pclass_2	Pclass_3	female	   male
    0	    0	        0	        1	       0	    1
    1	    1	        0	        0	       1	    0
    2	    0	        0	        1	       1	    0
    '''

    # Age
    dfresult['Age'] = dfdata['Age'].fillna(0) # Fill NA/NaN values using the specified method.
    dfresult['Age_null'] = pd.isna(dfdata['Age']).astype('int32') # Detect missing values for an array-like object.
    '''

        Pclass_1	Pclass_2	Pclass_3	female	    male	Age	    Age_null
    0	    0	    0	        1	           0	    1	    22.0	    0
    1	    1	    0	        0	           1	    0	    38.0	    0
    2	    0	    0	        1	           1	    0	    26.0	    0
    '''

    # Sibsp, Parch, Fare
    dfresult['SibSp'] = dfdata['SibSp']
    dfresult['Parch'] = dfdata['Parch']
    dfresult['Fare'] = dfdata['Fare']

    # Carbin
    dfresult['Cabin_null'] = pd.isna(dfdata['Cabin']).astype('int32')

    # Embarked
    dfEmbarked = pd.get_dummies(dfdata['Embarked'], dummy_na=True) # dummy_na=True: Add a column to indicate NaNs, if False NaNs are ignored.
    dfEmbarked.columns = ['Embarked_' + str(x) for x in dfEmbarked.columns]
    dfresult = pd.concat([dfresult, dfEmbarked], axis=1)

    return dfresult
```

```python
x = preprocessing(dftrain_raw).values
y = dftrain_raw[['Survived']].values

num = x.shape[0]
num_val = np.floor(num*0.2).astype('int32')

row_rand_array = np.arange(num)
np.random.shuffle(row_rand_array)

x_val = x[row_rand_array[:num_val], :]
y_val = y[row_rand_array[:num_val], :]

x_train = x[row_rand_array[num_test:], :]
y_train = y[row_rand_array[num_test:], :]

x_train.shape, y_train.shape, x_val.shape, y_val.shape
```

```python
dl_train = DataLoader(TensorDataset(torch.tensor(x_train).float(), torch.tensor(y_train).float()), shuffle=True, batch_size=8)
dl_valid = DataLoader(TensorDataset(torch.tensor(x_test).float(),torch.tensor(y_test).float()), shuffle=False, batch_size=8)
```

# 定义模型

使用Pytorch有三种构建模型:
- 使用nn.Sequential按层顺序构建模型
- 继承nn.Module基类构建模型
- 继承nn.Module基类构建模型并辅助应用模型容器进行封装

此处选择使用最简单的nn.Sequential,按层顺序模型

```python
def create_net():
    net = nn.Sequential()
    net.add_module("Linear1", nn.Linear(15, 20))
    net.add_module("relu1", nn.ReLU())
    net.add_module("Linear2", nn.Linear(20, 15))
    net.add_module("relu2", nn.ReLU())
    net.add_module("Linear3", nn.Linear(15, 1))
    net.add_module("sigmoid", nn.Sigmoid())
    return net

net = create_net()
print(net)
```

```python
from torchkeras import summary
summary(net, input_shape=(15, ))
```

# 三、训练模型
Pytorch需要用户自行编写自定义循环，有三种类型的训练循环代码风格：
- 脚本形式训练循环
- 函数形式训练循环
- 类形式训练循环

这里使用脚本形式。

```python
from sklearn.metrics import accuracy_score # Accuracy classification score.
 
loss_func = nn.BCELoss()
'''
torch.optim: https://pytorch.org/docs/stable/optim.html?highlight=torch%20optim#module-torch.optim

''' 
optimizer = torch.optim.Adam(params=net.parameters(), lr=0.01)
metric_func = lambda y_pred, y_true: accuracy_score(y_true.data.numpy(), y_pred.data.numpy()>0.5)
metric_name = "accuracy"
```

https://pytorch.org/docs/stable/generated/torch.nn.BCELoss.html?highlight=nn%20bceloss#torch.nn.BCELoss

BCELoss: A Criterion that measures the Binary Cross Entropy between the target and the output.

The unreduced (i.e. with reduction set to 'none') loss can be described as:

$l(x, y) = L = \{l_1, \cdots, l_N\}^T \; , l_n = -\omega_n [y_n \dot \log x_n + (1-y_n) \dot log(1-x_n)] $

```python
epochs = 10
log_step_freq = 30

dfhistory = pd.DataFrame(columns= ["epoch", "loss", metric_name, "val_loss", "val_" + metric_name])
print("Start Training...")
now_time = datetime.datetime.now().strftime('%Y-%m-%d %H:%M:%S')
print("==========="*8 + "%s"%now_time)

for epoch in range(1, epochs + 1):
    net.train()
    loss_sum = 0.0
    metric_sum = 0.0
    step = 1
    
    # 训练循环----------------------------------------------------------------------------------
    for step, (features, labels) in enumerate(dl_train, 1):
        
        # 梯度清零
        optimizer.zero_grad()
        
        # 正向传播求损失
        predictions = net(features) # 前向传播
        loss = loss_func(predictions, labels) # 计算一个batch的平均loss
        metric = metric_func(predictions, labels) # 一个batch训练后模型预测的精度
        
        # 反向传播求梯度
        loss.backward()
        optimizer.step()
        
        # 打印batch级别日志
        loss_sum += loss.item()
        metric_sum += metric.item()
        
        if step%log_step_freq == 0:
            print( ("[step = %d] loss: %.3f, " + metric_name + ": %.3f") %(step, loss_sum/step, metric_sum/step) )
    
    # 验证循环----------------------------------------------------------------------------------
    net.eval() # Set the Module in evaluation mode
    val_loss_sum = 0.0
    val_metric_sum = 0.0
    val_step = 1
    
    for val_step, (features, labels) in enumerate(dl_valid, 1): # enumerate()函数用于将一个可遍历的数据对象(如列表、元组或字符串)组合为一个索引序列
        with torch.no_grad():
            predictions = net(features)
            val_loss = loss_func(predictions, labels)
            val_metric = metric_func(predictions, labels)
        val_loss_sum += val_loss.item()
        val_metric_sum += val_metric.item()
        
    # 记录日志
    info = (epoch, loss_sum/step, metric_sum/step, val_loss_sum/val_step, val_metric_sum/val_step)
    dfhistory.loc[epoch - 1] = info
    
    print(("\nEPOCH = %d, loss = %.3f," + metric_name + \
           " = %.3f, val_loss = %.3f, " + "val_" + metric_name+"=%.3f") %info)
    now_time = datetime.datetime.now().strftime('%Y-%m-%d %H:%M:%S')
    print("==========="*8 + "%s"%now_time)

print('Finished Training...')
```

```python
%matplotlib inline
%config InlineBackend.figure_format = 'svg'

def plot_metric(dfhistory, metric):
    train_metrics = dfhistory[metric]
    val_metrics = dfhistory['val_' + metric]
    epochs = range(1, len(train_metrics) + 1)
    plt.plot(epochs, train_metrics, 'bo--')
    plt.plot(epochs, val_metrics, 'ro--')
    plt.title('Training and validation ' + metric)
    plt.xlabel('Epochs')
    plt.ylabel(metric)
    plt.legend(['train_' + metric, 'val_' + metric])
    plt.show()
```

```python
plot_metric(dfhistory, 'loss')
```

```python
plot_metric(dfhistory, 'accuracy')
```

# 六.保存模型

Pytorch 有两种保存模型的方式，都是通过调用pickle序列化方法实现的。

第一种方法只保存模型参数。

第二种方法保存完整模型。

推荐使用第一种，第二种方法可能在切换设备和目录的时候出现各种问题。

## 1.保存模型参数

```python
print(net.state_dict().keys()) # Returns a dictionary containing a whole state of the module.
```

```python
torch.save(net.state_dict(), './data/net_parameter.pkl')

# 创建同样的模型并加载参数
net_clone = create_net()
net_clone.load_state_dict(torch.load("./data/net_parameter.pkl"))
net(torch.tensor(x_tst[0:10]).float()).data
```
