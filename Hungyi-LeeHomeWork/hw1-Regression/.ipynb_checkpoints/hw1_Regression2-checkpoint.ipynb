## 数据预处理

```python
data_r = pd.read_csv('./hw1_data/train.csv', encoding='big5')
data_r[data_r == 'NR'] = 0
data_r.head(36)
```

```python
arr_r = np.array(data_r.loc[:,'0':'23']).astype(float)
arr_r.shape
```

```python
X = arr_r[np.newaxis,:18,:10]
X.shape
```

```python
training_data_size = arr_r.shape[0]/18
training_data_size= int(training_data_size)
training_data_size
```
