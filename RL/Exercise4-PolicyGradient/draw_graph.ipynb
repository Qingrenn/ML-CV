```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
scores = np.load("./data/scores.npy")
losses = np.load("./data/losses.npy")
scores_tst = np.load("./data/scores_tst.npy")
```

```python
scores.shape,losses.shape,scores_tst.shape
```

```python
def meanScores(scores):
    mean = []
    sum = 0
    for i in range(len(scores)):
        sum += scores[i]
        mean.append(sum/(i+1))
    return mean
```

```python
mean_scores = meanScores(scores)
```

```python
# 绘制mean_rewards-episodes
fig1 = plt.figure(figsize=(8,8))
ax1 = fig1.add_subplot(1,1,1)
ax1.plot(range(len(mean_scores)), mean_scores, 'blue')
ax1.set_xlabel("episodes", fontsize=15)
ax1.set_ylabel("scores", fontsize=15)
ax1.set_title("mean-rewards", fontsize=15)
fig1.savefig("./graph/mean_rewards.png")
```

```python
# 绘制loss
fig2 = plt.figure(figsize=(8,8))
ax2 = fig2.add_subplot(1,1,1)
ax2.plot(range(len(losses)), losses)
ax2.set_ylim(np.min(losses)-0.01, np.max(losses)+0.01)
ax2.set_xlabel("episodes", fontsize=15)
ax2.set_ylabel("losses", fontsize=15)
ax2.set_title("loss", fontsize=15)
fig2.savefig("./graph/loss.png")
```

```python
# 绘制训练得分
fig3 = plt.figure(figsize=(8,8))
ax3 = fig3.add_subplot(1,1,1)
ax3.plot(range(len(scores_tst)), scores_tst, 'blue')
ax3.set_xlabel("episodes", fontsize=15)
ax3.set_ylabel("scores", fontsize=15)
ax3.set_title("test scores", fontsize=15)
fig3.savefig("./graph/test-scores.png")
```
