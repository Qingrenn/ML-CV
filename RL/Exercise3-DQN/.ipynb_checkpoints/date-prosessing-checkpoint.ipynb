```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
scores = np.load('./data/scores.npy')
losses = np.load('./data/losses.npy')
```

```python
def meanScores(scores):
    mean = []
    sum = 0
    for i in range(len(scores)):
        sum += scores[i]
        mean.append(sum/(i+1))
    return mean
```

```python
mean_scores = meanScores(scores)
```

```python
plt.plot(range(len(mean_scores)), mean_scores,'blue')
plt.xlabel('episodes')
plt.ylabel('rewards')
plt.title('mean rewards')
plt.savefig('./graph/mean_rewards')
```

```python
plt.plot(range(len(losses)), losses,'blue')
plt.xlabel('episodes')
plt.ylabel('losses')
plt.title('Loss')
plt.savefig('./graph/losses.png')
```
